```python
df['Age'].fillna(round(df['Age'].mean(),1),inplace=True)
df["Cabin"].fillna(value='Not Known',inplace=True)
df["Embarked"].fillna(df["Embarked"].mode(),inplace=True)
df
```

# Check and handle outliers in at least 3 columns in the datas

# OUTLIERS IN FARE

```python
plt.boxplot(df['Fare'])
plt.title('box plot of Fare')
```

```python
Q1=np.percentile(df['Fare'],25,interpolation='midpoint')
Q2=np.percentile(df['Fare'],50,interpolation='midpoint')
Q3=np.percentile(df['Fare'],75,interpolation='midpoint')
print('Q1=',Q1)
print('Q2=',Q2)
print('Q3=',Q3)

```

```python
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
upp_lim=Q3+1.5*IQR
print(low_lim)
print(upp_lim)
```

```python
outlier=[]
for x in df['Fare']:
    if((x>upp_lim) or (x<low_lim)):
        outlier.append(x)
outlier
```

```python
df['Fare']=np.where(df['Fare']>upp_lim,upp_lim,np.where(df['Fare']<low_lim,low_lim,df['Fare']))
df
```

```python
plt.boxplot(df['Fare'])
plt.title('box plot of Fare')
```
